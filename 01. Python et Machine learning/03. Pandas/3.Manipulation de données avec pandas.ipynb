# Manipulation de données avec la bibliothèque Pandas

- Importer la librairie pandas.
- Lire le fichier "food_info.csv" dans un dataframe qu'on nommera food_info.
- Utiliser l'attribut Dataframe.columns suivi de la méthode Index.tolist() pour retourner une liste contenant tous les noms de colonne de food_info.
- Assigner la liste résultante à la variable col_names et utiliser la fonction print() pour afficher le nom des colonnes.
- Enfin afficher les 3 premiers éléments de food_info.

```python
import pandas as pd

food_info = pd.read_csv("Jeux_de_donnees/food_info.csv")
col_names = food_info.columns.tolist()
```

```python
print(col_names)
```

```python
food_info.head(3)
```

## Transformer une colonne

- Diviser la colonne "Sodium_(mg)" par 1000 pour convertir cette colonne en grammes et assigner le résultat à la variable sodium_grams.
- Multiplier la colonne "Sugar__Tot_(g)" par 1000 pour convertir en milligrammes et assigner le résultat à la variable sugar_milligrams.

```python
sodium_grams = food_info["Sodium_(mg)"] / 1000
sugar_milligrams = food_info["Sugar_Tot_(g)"] * 1000
```

## Opérations mathématiques entre colonnes

- Diviser la colonne "Protein_(g)" par la colonne "Water_(g)" et assigner le résultat à la variable grams_of_protein_per_gram_of_water.
- Additionner les colonnes "Calcium_(mg)" et "Iron_(mg)" et assigner le résultat à la variable milligrams_of_calcium_and_iron.
- Afficher les résultats.

```python
grams_of_protein_per_gram_of_water = food_info["Protein_(g)"] / food_info["Water_(g)"]
milligrams_of_calcium_and_iron = food_info["Calcium_(mg)"] + food_info["Iron_(mg)"]
```

```python
print(grams_of_protein_per_gram_of_water)
```

```python
print(milligrams_of_calcium_and_iron)
```

## Créer un indice nutritionnel

Score = 2 x (Protein_(g)) - 0.75 x (Lipid_Tot_(g))

- Multiplier la colonne "Protein_(g)" par 2 et assigner le résultat à la variable protein.
- Multiplier la colonne "Lipid_Tot_(g)" par -0.75 et assigner le résultat à la variable fat.
- Additionner les variables ensemble et assigner le résultat à la variable rating.
- Afficher le résultat.

```python
protein = food_info["Protein_(g)"] * 2
fat = food_info["Lipid_Tot_(g)"] * (0.75)
rating = protein - fat
print(rating)
```

## Normaliser des colonnes

- Normaliser les valeurs de la colonne "protein_(g)" et assigner le résultat à la variable normalized_protein.
- Normaliser les valeurs de la colonne "Lipid_Tot_(g)" et assigner le résultat à la variable normalized_fat.

```python
normalized_protein = food_info["Protein_(g)"] / food_info["Protein_(g)"].max()
normalized_fat = food_info["Lipid_Tot_(g)"] / food_info["Lipid_Tot_(g)"].max()
```

```python
print(normalized_fat)
```

## Créer une nouvelle colonne

- Assigner la colonne normalisée représentée par la variable normalized_protein à la nouvelle colonne qu'on nommera "Normalized_Protein" dans le dataframe food_info.
- Assigner la colonne normalisée représentée par la variable normalized_fat à la nouvelle colonne qu'on nommera "Normalized_Fat" dans la dataframe food_info.
- Afficher les 5 premières valeurs du dataframe food_info.

```python
food_info["Normalized_Protein"] = normalized_protein
food_info["Normalized_Fat"] = normalized_fat
```

```python
food_info.head()
```

## Créer une colonne pour l'indice nutritionnel normalisé

Score = 2 x Normalized_Protein - 0.75 x Normalized_Fat

- Utiliser les colonnes Normalized_Protein et Normalized_Fat avec cette formule pour créer une nouvelle colonne qu'on nommera Norm_Nutr_Index.

```python
food_info["Norm_Nutr_Index"] = 2 * food_info["Normalized_Protein"] + (-0.75)*food_info["Normalized_Fat"]
```

```python
food_info.head()
```

## Trier un DataFrame

- Remplacer le dataframe food_info en triant par rapport à la colonne Norm_Nutr_Index de façon décroissante (du plus grand au plus petit)

```python
food_info.sort_values("Norm_Nutr_Index", inplace=True, ascending=False)
```
